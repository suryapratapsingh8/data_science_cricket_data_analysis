```python
import pandas as pd
import json
```

```python
with open("/content/t20_wc_match_results.json") as f:
  data = json.load(f)
df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.shape
```

```python
df_match.rename(columns={'scorecard':'match_id'}, inplace=True)
df_match.head()
```

```python
with open('/content/t20_wc_batting_summary.json') as f:
  data=json.load(f)
  all_records=[]
  for rec in data:
    all_records.extend(rec['battingSummary'])
df_batting=pd.DataFrame(all_records)
df_batting.head()
```

```python
df_batting['out/not_out']=df_batting.dismissal.apply(lambda x:'out' if len(x)>0 else 'not_out')
df_batting.head(20)
```

```python
df_batting.drop(columns=['dismissal'],inplace=True)
df_batting.head(10)
```

```python
df_batting['batsmanName']=df_batting['batsmanName'].apply(lambda x: x.replace(''," "))
df_batting['batsmanName']=df_batting['batsmanName'].apply(lambda x: x.replace('\a0'," "))
df_batting.head(11)
```

```python
match_ids={}
for index,row in df_match.iterrows():
  key1=row['team1']+' Vs '+row['team2']
  key2=row['team2']+' Vs '+row['team1']
  match_ids[key1]=row['match_id']
  match_ids[key2]=row['match_id']
match_ids
```

```python
match_ids['England Vs India']
```

```python
df_batting['match_id']=df_batting['match'].map(match_ids)
df_batting.head()
```

```python
print(df_batting.to_csv('Batting_stats.csv',index=False))
```

```python
df_match.to_csv('Match_summary.csv',index=False)
```

```python
with open('/content/t20_wc_bowling_summary.json') as f:
  data=json.load(f)
  all_records=[]
  for rec in data:
    all_records.extend(rec['bowlingSummary'])

df_bowling=pd.DataFrame(all_records)
df_bowling.head()
```

```python
df_bowling['match_id']=df_bowling['match'].map(match_ids)
df_bowling.head()
```

```python
df_bowling.to_csv("Bowling_stats.csv",index=False)
```
